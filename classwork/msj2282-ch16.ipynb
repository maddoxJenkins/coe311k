```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return -2 * y
def exact_sol(x):
    return 1 * np.exp(-2 * x)
x_start, x_end = 0, 10
y0 = 1 
h_values = [0.1, 0.5, 1.0, 1.5] 
plt.figure(figsize=(12, 8))
for h in h_values:
    num_steps = int((x_end - x_start) / h)
    x_vals = np.linspace(x_start, x_end, num_steps + 1)
    y_euler = np.zeros(num_steps + 1)
    y_euler[0] = y0
    for i in range(num_steps):
        y_euler[i+1] = y_euler[i] + h * f(x_vals[i], y_euler[i]) 
    y_true = exact_sol(x_vals)
    global_error = np.abs(y_true - y_euler)
    plt.plot(x_vals, y_euler, '-o', label=f'Euler h={h}')
x_fine = np.linspace(0, 10, 500)
plt.plot(x_fine, exact_sol(x_fine), 'k--', label='Exact Solution', linewidth=2)
plt.title("Euler's Method Stability")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
