```python
import numpy as np
import time
import matplotlib.pyplot as plt
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0]*n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C
matrix_indices = list(range(1, 9))
naive_runtimes = []
numpy_runtimes = []
rows, cols = 2, 3
for i in matrix_indices:
    size = rows if rows == cols else rows 
    A = np.random.rand(size, size)
    B = np.random.rand(size, size)
    start = time.time()
    _ = naive_matrix_multiplication(A.tolist(), B.tolist())
    naive_runtimes.append(time.time() - start)
    start = time.time()
    _ = np.dot(A, B)
    numpy_runtimes.append(time.time() - start)
    rows *= 2
    cols *= 2
plt.figure(figsize=(10, 6))
plt.plot(matrix_indices, naive_runtimes, label='o n^3', marker='o')
plt.plot(matrix_indices, numpy_runtimes, label='np.dot', marker='s')
plt.xlabel('Matrix Pair Number')
plt.ylabel('Runtime')
plt.title('Runtime Comparison Naive vs. Optimized Matrix Multiplication')
plt.legend()
plt.show()
```
