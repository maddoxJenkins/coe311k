```python
import numpy as np
import matplotlib.pyplot as plt

a, b = 0, 2 * np.pi
N = 50
x = np.linspace(a, b, N)
f = lambda x: np.sin(x)
h = (b - a) / (N - 1) 
df_central = (f(x + h) - f(x - h)) / (2 * h)
df_forward = (f(x + h) - f(x)) / h 
df_backward = (f(x) - f(x - h)) / h 

fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 0].plot(x, np.cos(x), color='black')
axs[0, 0].set_title("cos(x)")
axs[0, 1].plot(x, np.cos(x), 'k--', alpha=0.3, label='Exact')
axs[0, 1].plot(x, df_forward, color='red', label='Forward')
axs[0, 1].set_title("Forward Difference")
axs[0, 1].legend()
axs[1, 0].plot(x, np.cos(x), 'k--', alpha=0.3, label='Exact')
axs[1, 0].plot(x, df_backward, color='green', label='Backward')
axs[1, 0].set_title("Backward Difference")
axs[1, 0].legend()
axs[1, 1].plot(x, np.cos(x), 'k--', alpha=0.3, label='Exact')
axs[1, 1].plot(x, df_central, color='blue', label='Central')
axs[1, 1].set_title("Central Difference")
axs[1, 1].legend()
plt.tight_layout()
plt.show()
```
