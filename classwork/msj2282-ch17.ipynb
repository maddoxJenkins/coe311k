```python
import numpy as np
import matplotlib.pyplot as plt

k = 1.0 
x_start = 0
x_end = 10
y0 = 1.0 
v0 = 0.0 
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]
plt.figure(figsize=(12, 8))
for h in h_values:
    x_vals = np.arange(x_start, x_end + h, h)
    y_vals = np.zeros(len(x_vals))
    v_vals = np.zeros(len(x_vals))
    y_vals[0] = y0
    v_vals[0] = v0 
    for i in range(1, len(x_vals)):
        y_n = y_vals[i-1]
        v_n = v_vals[i-1]
        y_vals[i] = y_n + h * v_n
        v_vals[i] = v_n + h * (-k * y_n)
    plt.plot(x_vals, y_vals, label=f'h = {h}')
plt.title("Stability of Euler's Method")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

Unstabl3 when h inscreases, especially when h reaches around .5 and has more extreme oscilations. When k = 10, the frequency of oscillation increases significantly making it very unstable. Eulers method can not solve this because the error is proportional to h.
