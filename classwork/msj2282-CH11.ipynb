```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])
cs = CubicSpline(x, y, bc_type='natural')
poly_coeffs = np.polyfit(x, y, 5)
poly_func = np.poly1d(poly_coeffs)
x_interp = np.linspace(0, 5, 100)
y_spline = cs(x_interp)
y_poly = poly_func(x_interp)
plt.figure(figsize=(10, 6))
plt.plot(x_interp, y_spline, 'r-', label="Cubic Spline")
plt.plot(x_interp, y_poly, 'g--', label="5th-Degree Polynomial")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
