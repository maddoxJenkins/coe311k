```python
import numpy as np
X = np.sort(np.random.rand(50) * 10) 
Y = np.random.rand(50) * 10
A = np.vstack([X, np.ones(len(X))]).T
sol = np.linalg.lstsq(A, Y, rcond=None)[0]
m, c = sol
print(f"Slope (m): {m}, Intercept (c): {c}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
m_orig = 2.5
b_orig = 1.2
X = np.arange(0, 1.01, 0.01)
noise = np.random.rand(len(X))
Y = m_orig * X + b_orig + noise
A = np.vstack([X, np.ones(len(X))]).T
m_pred, c_pred = np.linalg.lstsq(A, Y, rcond=None)[0]
Y_pred = m_pred * X + c_pred
residuals = Y - Y_pred
max_idx = np.argmax(np.abs(residuals))
min_idx = np.argmin(np.abs(residuals))
print(f"Original: m={m_orig}, b={b_orig}")
print(f"Predicted: m={m_pred:.4f}, c={c_pred:.4f}")
print(f"Point with largest error: X={X[max_idx]:.2f}, Residual={residuals[max_idx]:.4f}")
print(f"Point with smallest error: X={X[min_idx]:.2f}, Residual={residuals[min_idx]:.4f}")
plt.scatter(X, Y, label='Observed data points', color='blue', alpha=0.5)
plt.plot(X, Y_pred, label='Least squares best-fit line', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Challenge: Linear Least Squares Fit')
plt.legend()
plt.show()
```
