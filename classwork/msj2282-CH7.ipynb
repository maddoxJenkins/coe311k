```python
import numpy as np
import time

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    start_time = time.time()
    
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
            if np.linalg.norm(x_new - x, ord=np.inf) < tol:
                return x_new, k + 1, time.time() - start_time
        x = x_new
        
    return x, max_iterations, time.time() - start_time

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    start_time = time.time()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
            
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k + 1, time.time() - start_time
        x = x_new
    return x, max_iterations, time.time() - start_time

tol = 1e-6
max_iterations = 100000


A1 = np.array([
    [0.582745, 0.48, 0.10, 0., 0.],
    [0.48, 1.044129, 0.46, 0.10, 0.],
    [0.10, 0.46, 1.10431, 0.44, 0.10],
    [0., 0.10, 0.44, 0.963889, 0.42],
    [0., 0., 0.10, 0.42, 0.522565]], dtype=float)
b1 = np.array([1.162745, 2.084129, 2.20431, 1.923889, 1.042565], dtype=float)

A2 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b2 = np.array([1, 1], dtype=float)

A3 = np.array([[1, 1], [1, 1.0001]], dtype=float)
b3 = np.array([1, 1.0001], dtype=float)

def solve_system(name, A, b):
    x0 = np.zeros_like(b)
    j_sol, j_iter, _ = jacobi(A, b, x0, tol, max_iterations)
    gs_sol, gs_iter, _ = gauss_seidel(A, b, x0, tol, max_iterations)
    print(f"Jacobi: {j_iter} iterations | Result: {j_sol}")
    print(f"Gauss-Seidel: {gs_iter} iterations | Result: {gs_sol}\n")

solve_system("System 1", A1, b1)
solve_system("System 2", A2, b2)
solve_system("System 3", A3, b3)
```
