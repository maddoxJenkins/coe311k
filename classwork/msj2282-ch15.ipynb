```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return -2 * y 
x_start, x_end = 0, 2 
step_sizes = np.arange(0.01, 1.05, 0.05) 
baseline_y = None
baseline_x = None
plt.figure(figsize=(10, 6))
for h in step_sizes:
    num_steps = int((x_end - x_start) / h) 
    x_values = np.linspace(x_start, x_end, num_steps + 1) 
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1 
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i]) 
    plt.plot(x_values, y_values, label=f"h={h:.2f}")
    if baseline_y is None:
        baseline_y = y_values
        baseline_x = x_values
        print(f"Run h={h:.2f}: Baseline set.")
    else:
        indices = np.searchsorted(baseline_x, x_values)
        current_errors = np.abs(y_values - baseline_y[indices])
        total_error = np.sum(current_errors)
        print(f"Run h={h:.2f} | Total Accumulated Error vs Baseline: {total_error:.4f}")
plt.xlabel('x') 
plt.ylabel('y')
plt.title("Euler's Method: Impact of Step Size (h) on Error") 
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show() 
```
