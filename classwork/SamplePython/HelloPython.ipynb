```python
import math

def calc_hyp(a,b):
    h = math.sqrt(a*a+ b*b)
    return h
```

```python
import math
import random
def approximatePi(numSamples):
    counter = 0;
    for i in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(x*x+y*y<1):
            counter+=1
    return 4.0*counter/numSamples

        
```

```python
print(approximatePi(10000))
```
