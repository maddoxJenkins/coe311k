```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
def approximatePi(numSamples):
    counter = 0;
    for i in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(x*x+y*y<1):
            counter+=1
    return 4.0*counter/numSamples
arra = ([])
arrr = ([])
x = 1
for x in range(100):
    arra = np.append(arra, abs(approximatePi(1000)-np.pi))
    arrr = np.append(arrr, abs((approximatePi(1000)-np.pi)/np.pi))
x = np.arange(1,101,1)
plt.plot(x, arra, label='true error')
plt.plot(x, arrr, label='relative error')
plt.legend()
plt.ylabel("error value")
plt.xlabel("sample number")
plt.show()
```
