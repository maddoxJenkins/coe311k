```python
import numpy as np
import matplotlib.pyplot as plt
m_orig = 2.5
b_orig = 1.2
X = np.arange(0, 1.01, 0.01)
noise = np.random.rand(len(X)) 
Y = m_orig * X + b_orig + noise
A = np.vstack([X, np.ones(len(X))]).T
A_transpose = A.T
beta = np.linalg.inv(A_transpose @ A) @ A_transpose @ Y
m_custom, b_custom = beta
m_np, b_np = np.linalg.lstsq(A, Y, rcond=None)[0]
Y_pred = m_custom * X + b_custom
residuals = Y - Y_pred
max_idx = np.argmax(np.abs(residuals))
min_idx = np.argmin(np.abs(residuals))
print(f"Original Equation: y = {m_orig}x + {b_orig}")
print(f"Custom Solver:  m = {m_custom:.4f}, b = {b_custom:.4f}")
print(f"NumPy Solver:   m = {m_np:.4f}, b = {b_np:.4f}")
print(f"Max Residual: {residuals[max_idx]:.4f} at X = {X[max_idx]:.2f}")
print(f"Min Residual: {residuals[min_idx]:.4f} at X = {X[min_idx]:.2f}")
plt.figure(figsize=(10, 6))
plt.scatter(X, Y)
plt.plot(X, Y_pred, color = 'red')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
