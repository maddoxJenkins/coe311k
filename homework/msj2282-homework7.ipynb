```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return -2 * y * np.cos(x)
def exact_sol(x):
    return np.exp(-2 * np.sin(x))
x_start = 0
x_end = 10 * np.pi
y0 = 1
h_values = [0.01, 0.05, 0.15, 0.2, 0.25]
results = {}
avg_errors = []
local_errors_at_15 = []
for h in h_values:
    x_vals = np.arange(x_start, x_end + h, h)
    y_vals = np.zeros(len(x_vals))
    y_vals[0] = y0
    for i in range(len(x_vals) - 1):
        y_vals[i+1] = y_vals[i] + h * f(x_vals[i], y_vals[i])
    results[h] = (x_vals, y_vals)
    y_exact = exact_sol(x_vals)
    errors = np.abs(y_exact - y_vals)
    avg_errors.append(np.mean(errors))
    idx_15 = (np.abs(x_vals - 15)).argmin()
    local_errors_at_15.append(np.abs(exact_sol(x_vals[idx_15]) - y_vals[idx_15]))
plt.figure(figsize=(10, 5))
for h in h_values:
    plt.plot(results[h][0], results[h][1], label=f'h = {h}')
x_fine = np.linspace(x_start, x_end, 1000)
plt.plot(x_fine, exact_sol(x_fine), 'k--', alpha=0.5)
plt.title("Euler's Method")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(h_values, avg_errors, 'o-', color='red')
plt.title("Global Error")
plt.xlabel('h')
plt.ylabel('Average Error')
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(h_values, local_errors_at_15, 's-', color='green')
plt.title("Local Error")
plt.xlabel('h')
plt.ylabel('Local Error')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x, y):
    return -2 * y * np.cos(x) 
def exact_sol(x):
    return np.exp(-2 * np.sin(x)) 
x_start = 0
x_end = 10 * np.pi
y0 = 1
h_test_values = [0.1, 0.5, 1.0, 1.1] 
plt.figure(figsize=(12, 6))
for h in h_test_values:
    num_steps = int((x_end - x_start) / h)
    x_vals = np.linspace(x_start, x_end, num_steps + 1)
    y_vals = np.zeros(num_steps + 1)
    y_vals[0] = y0
    for i in range(num_steps):
        y_vals[i+1] = y_vals[i] + h * f(x_vals[i], y_vals[i])
    plt.plot(x_vals, y_vals, label=f'h = {h}')
x_fine = np.linspace(x_start, x_end, 1000)
plt.plot(x_fine, exact_sol(x_fine), 'k--', alpha=0.6, label='Exact Solution')
plt.ylim(-2, 5) 
plt.title("Stability Analysis")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
