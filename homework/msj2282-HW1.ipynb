```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
def approximatePi(numSamples):
    counter = 0;
    for i in range(0, numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(x*x+y*y<1):
            counter+=1
    return 4.0*counter/numSamples
arr = np.arange(10, 10001, 20)
plot = np.array([])
for x in arr:
    plot = np.append(plot, abs(approximatePi(x)-np.pi))
plt.plot(arr, plot)
plt.xlabel('samples')
plt.ylabel('error')
plt.show()
```
