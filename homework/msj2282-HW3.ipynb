```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
def factorial(x):
    if x <= 1: 
        return 1
    return x * factorial(x-1)
def taylorsin(n,y):
    sum = 0
    for x in range(0,n+1,1):
        sum+=((-1)**x)*((y**(2*x+1))/(factorial(2*x+1)))
    return sum
arr = np.arange(0,10.1,.1)
error1 = np.array([])
error2 = np.array([])
error3 = np.array([])
error4 = np.array([])
for x in arr:
    error1 = np.append(error1, abs(taylorsin(1,x)- np.sin(x)))
    error2 = np.append(error2, abs(taylorsin(2,x)- np.sin(x)))
    error3 = np.append(error3, abs(taylorsin(3,x)- np.sin(x)))
    error4 = np.append(error4, abs(taylorsin(4,x)- np.sin(x)))
plt.plot(arr, error1, label = '1 term')
plt.plot(arr, error2, label = '2 terms')
plt.plot(arr, error3, label = '3 terms')
plt.plot(arr, error4, label = '4 terms')
plt.xlabel('x-value')
plt.ylabel('error')
plt.legend()
plt.show()
```

As you increase the number of terms, the error grows exponentially. The error does not decrease uniformly for all values of x as 1 term does a squiggle if you examine it in a vaccum and zoom in. When a large value is passed through, the 4 term error is so much greater than any other amount of terms. The main conclusion is that more terms are accurate for very small x values but very inaccurate for larger x values.
